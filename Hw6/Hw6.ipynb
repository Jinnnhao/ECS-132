```r
# Define the range for x*
x_star_range <- seq(52, 65, length.out = 100)

# Mean and standard deviation for the distributions
mean_D0 <- 50
std_D0 <- 10
mean_D1 <- 70
std_D1 <- 15

# Calculate TPR and FPR for each value of x*
TPR <- 1 - pnorm(x_star_range, mean = mean_D1, sd = std_D1)
FPR <- 1 - pnorm(x_star_range, mean = mean_D0, sd = std_D0)

# Plot the ROC curve
plot(FPR, TPR, type = 'l', col = 'blue', lwd = 2, xlab = 'False Positive Rate (FPR)', ylab = 'True Positive Rate (TPR)', main = 'ROC Curve')
abline(0, 1, col = 'red', lty = 2) # Add diagonal line representing random guessing
grid() # Add grid
```
