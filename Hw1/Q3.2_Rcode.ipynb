```r
set.seed(123)
total_songs = 100 
shuffle_times = 100000
plotprob = numeric(shuffle_times)
bingo = 0
for(k in 1:shuffle_times){
    shuffled_songs = sample(1:100, total_songs, replace = FALSE) 
    # lets say cake's song in range of [1,10]  
    if((1 <= shuffled_songs[5] && shuffled_songs[5] <= 10) && shuffled_songs[1] > 10 && shuffled_songs[2] > 10 && shuffled_songs[3] > 10 && shuffled_songs[4] > 10){
     bingo = bingo + 1  # record if it meets the condition
    } 
    plotprob[k] = bingo / k
}
overall_prob = bingo / shuffle_times
print(overall_prob)
plot(plotprob, ylab="Probability", xlab="Number of shuffle times (k)")

```
