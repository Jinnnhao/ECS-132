Since discarding those that don't work.
The probability that the first key failure to open the door is:
\begin{equation*}
  \frac{n-1}{n}
\end{equation*}
The probability that the second key failure to open the door is:
\begin{equation*}
  \frac{n-2}{n-1}
\end{equation*}
The probability that the $kth$ key works is:
\begin{equation*}
  \frac{1}{n-k+1}
\end{equation*}
Thus, the overall probability that the $kth$ key works is:
\begin{equation*}
P_k = \frac{n-1}{n} \times  \frac{n-2}{n-1} \times ...... \times \frac{1}{n-k+1} = \frac{1}{n}
\end{equation*}

## 
If doesn't discard the not working tried keys, 
So similarly, the probability that the first key faliure to open the door is: 
\begin{equation*}
\frac{n-1}{n}
\end{equation*}
The probability that the second key opens the door is:
\begin{equation*}
\frac{n-1}{n}
\end{equation*}
Thus, the $kth$ keys that opens the door is:
\begin{equation*}
\frac{1}{n}
\end{equation*}
Thus, the overall probability that the $kth$ key works is:
\begin{eqnarray*}
P_k &=& \frac{n-1}{n} \times \frac{n-1}{n} \times...... \times \frac{1}{n} \\
    &=& (\frac{n-1}{n})^{k-1} \times \frac{1}{n}
\end{eqnarray*}

The total combinations of 10 red requests out of total 20 requests: 
\begin{eqnarray*}
C(20,10) &=& \binom{20}{10} \\
         &=& \frac{20!}{10! \times 10!} \\
         &=& 184756
\end{eqnarray*}
Then, each of the rest of 10 requests could be green or black, which has total situations: 
\begin{equation*}
S_1 = 2^{10}
\end{equation*}

Thus, the overall total request situations for each request could be red or green or black:
\begin{equation*}
S_0 = 3^{20}
\end{equation*}
So, the probability that receiving 10 red among of first 20 requests is:
\begin{eqnarray*}
P &=& \frac{\binom{20}{10} \times 2^{10}}{3^{20}} \\
  &=& \frac{184756 \times 2^{10}}{3^{20}} \\
  &=& 0.0543
\end{eqnarray*}

##  
Since each sensor is independent to each other.

Thus is time slot $i$, the probability that both sensors are active is:
\begin{equation*}
P_{(A_i)} = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}
\end{equation*}

## 
The probability that both sensors are active during two time slot:
\begin{eqnarray*}
P(A_i \cap A_j) &=& P(A_1) \times P(A_2) + P(A_1) \times P(A_3) + P(A_2) \times P(A_3) \\
                &=& \frac{1}{4} \times \frac{1}{4} + \frac{1}{4} \times \frac{1}{4} + \frac{1}{4} \times \frac{1}{4} \\
                &=& \frac {3}{16}
\end{eqnarray*}

## 
The probability that all senors are active during all three time slot is:
\begin{eqnarray*}
P(A_1 \cap A_2 \cap A_3) &=& \frac{1}{4} \times \frac{1}{4} \times \frac{1}{4} \\
                         &=& \frac{1}{64}
\end{eqnarray*}

## 
The probability that both sensors are active during at least one time slot is: 
\begin{eqnarray*}
P(A_1 \cup A_2 \cup A_3) &=& 1 - P(\text{None time slot that both sensor are active}) \\
                         &=& 1 - (\frac{3}{4})^3 \\
                         &=& \frac{37}{64}
\end{eqnarray*}

```r

n = 100
experiment = 1000 
prob = numeric(n)
for (num in 1:n){
    bingo = 0
    for (k in 1:experiment)
        {
            shuffled_hat = sample(1:num, num, replace = FALSE)
            for (i in 1:num)
                {
                    if (shuffled_hat[i] == i)
                        {
                            bingo = bingo + 1
                            break;
                        }
                }
        }
        prob[num] = bingo/experiment
}
plot(prob, ylab="Probability", xlab="Number of people (n)")
abline(h = 1 - 1/exp(1) , col = "red", lwd = 3)
```
