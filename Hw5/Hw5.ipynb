The probability that a node being active is P = 0.4

Then, the probability that a node is **not** being active is P = 1 - 0.4 = 0.6 

1. P(I) = the probability that 0 node is active:
   \begin{eqnarray*}
   P(I) &=& 0.6^{6} \\
        &=& 0.0467
   \end{eqnarray*}

   P(S) = the probability that only 1 node is active, out of 6 nodes. That means it belongs to Binomial distribution
   \begin{eqnarray*}
   P(I) &=& \binom{6}{1} \times 0.4^{1} \times 0.6^{5} \\
        &=& 6 \times 0.4^{1} \times 0.6^{5} \\
        &=& 0.187
   \end{eqnarray*}
2. Since *P(E) = 0.5*, *P(I|E) = 0.2*, *P(S|E) = 0.3*

   Thus, the probability the slot will be an collision slot given that the eavesdropper is listening:
   \begin{eqnarray*}
   P(C|E) &=& 1 - P(I|E) - P(S|E)\\
          &=& \frac{P(C \cap E)}{P(E)} = 0.5 \\ 
   \rightarrow P(C \cap E) &=& 0.5 \times P(E)\\
                  &=& 0.5 \times 0.5 \\
                  &=& 0.25
   \end{eqnarray*}

   And, the probability that the time slot is collision is P(C):
   \begin{eqnarray*}
   P(C) &=& 1 - P(I) - P(S) \\
        &=& 1 - 0.0467 - 0.187 \\
        &=& 0.766
   \end{eqnarray*}

   Thus, the probability that the eavesdropper is listening given that the time slot is a collision:
   \begin{eqnarray*}
    P(E|C) &=& \frac{P(E \cap C)}{P(C)} \\
           &=& \frac{P(C \cap E)}{P(C)} \\
           &=& \frac{0.25}{0.766}\\
           &=& 0.326
   \end{eqnarray*}

1. The CDF for k <= 10 is 0.

   Then, the CDF for k > 10 is:
 \begin{eqnarray*}
F_X (k) &=& P(X \leq k) \\
        &=& \int_{10}^k f(x) dx \\
        &=& \int_{10}^k \frac{10}{x^2} dx \\
        &=& 1 - \frac{10}{k}
\end{eqnarray*}

2. The probability that the device will fail within the first 15 hours, which mean $k = 15$
   
   \begin{eqnarray*}
    P(X \leq 15) &=& 1 - \frac{10}{15} \\
                 &=& 0.333 \\
   \end{eqnarray*}

3. Now we know that the probability that the device will fail within the first 15 hours $ P(fail)$ = $\frac{1}{3}$

   Thus, the probability that the device will **not** fail within the first 15 hours $ P(works) $= 1 - $\frac{1}{3}$ = $\frac{2}{3}$

   The probability that the $4th$ device will be the **first one** that **does not** fail within the first 15 hours, which belongs to Geometric distribution.

   So, k = 4.

   \begin{eqnarray*}
    P(X = k = 4) &=& (\frac{1}{3})^{k-1} \times \frac{2}{3} \\
                 &=& (\frac{1}{3})^{3} \times \frac{2}{3} \\
                 &=& \frac{2}{81} \\
                 &=& 0.0247
   \end{eqnarray*}

According to the problem, $\lambda_A$ = $\frac{1}{10}$, $\lambda_B$ = $\frac{1}{5}$

The probability that have to wait more than 8 mins for a bus to arrive, t = 8: 
\begin{eqnarray*}
P(X > 8) &=& P(X_A > 8 \cap X_B>8)\\
           &=& (1 - P_A(X \leq 8)) \times (1 - P_B(X \leq 8))\\
           &=& (1 - F_A(8)) \times (1 - F_B(8)) \\
           &=& (1 - (1 - e^{-\lambda_A t})) \times (1 - (1 - e^{-\lambda_B t}))\\
           &=&  e^{-\frac{1}{10} 8} \times e^{-\frac{1}{5} 8}\\
           &=& 0.0907
\end{eqnarray*}

1. In this problem, the mean $\mu = 71$ GBytes and the standard deviation $\sigma = 2.5$ GBytes. 

Thus, map it to stander normal distribution: $N(0,1)$,  $Z = \frac{X-\mu}{\sigma}$.
\begin{eqnarray*}
Z(72) &=& \frac{X-\mu}{\sigma}  \\
  &=& \frac{72 - 71}{2.5} \\
  &=& 0.4
\end{eqnarray*}

By using the Z table, the probability that the flows are greater than 72 GBytes: $P(X > 72) = 1 - P(X \leq 72)$
\begin{eqnarray*}
P(X > 72) &=& 1 - P(X \leq 72) \\
          &=& 1 - P(Z \leq Z(72)) \\
          &=& 1 - \Phi(Z(72)) \\
          &=& 1 - \Phi(0.4) \\
          &=& 1 - 0.6554 \\
          &=& 0.3446 \\
          &=& 34.46 \%
\end{eqnarray*}

2. Given P(X < m) = 88.30%, find m.
   Mapping it to stander normal distribution: $N(0,1)$,  $Z = \frac{X-\mu}{\sigma}$.
   \begin{eqnarray*}
    Z(m) &=& \frac{X-\mu}{\sigma}  \\
         &=& \frac{m - 71}{2.5} \\
   \end{eqnarray*}
   Thus,
   \begin{eqnarray*}
   P(X < m) &=& P(Z < Z(m)) \\
            &=& P(Z < \frac{m - 71}{2.5}) \\
            &=& \Phi(\frac{m - 71}{2.5}) \\
            &=& 0.8830 \\
   \end{eqnarray*}
   According the Z-table, $\Phi$(1.19) = 0.8830. Thus,
   \begin{eqnarray*}
   \frac{m - 71}{2.5} &=& 1.19 \\
                   m &=& 73.975
   \end{eqnarray*}
