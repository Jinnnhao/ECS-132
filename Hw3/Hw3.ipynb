Let:
Event $A$ = the probability that an email is spam = 80%

Thus the probability that an email is non-spam = 20%

Event $B$ = the probability that an email contains the phrase 
\begin{eqnarray*}
P(B)= 80\% \times 10\% + 20\% \times 1\%
\end{eqnarray*}

The probability that a spam email contain the phrase “Large inheritance” = 10%
\begin{eqnarray*}
P(B|A)= 10 \%
\end{eqnarray*}
Thus, the probability $P$ that a new email is received containing the phrase “Large inheritance” is spam is:

\begin{eqnarray*}
P = P(A|B) &=& \frac{P(A)\times P(B|A)}{P(B)}\\
           &=& \frac{80\% \times 10\%}{80\% \times 10\% + 20\% \times 1\%} = 97.6\%
\end{eqnarray*}

The probability that **correctly** receiving 0 is 0.94

The probability that **Not correctly receiving** 0 is 1-0.94 = 0.06

The probability that **correctly** receiving 1 is 0.91

The probability that **Not correctly receiving** 1 is 1-0.91 = 0.09

The probability that transmit 0 is 0.45, then the probability that transmit 1 is 0.55

1. Probability that a 1 is received. (It could be correct or an error.)
\begin{eqnarray*}
P_1 &=& 0.55 \times 0.91 + 0.45 \times 0.06 \\
  &=& 0.5275
\end{eqnarray*}

2. Probability that a 0 is received. (It could be correct or an error.)
\begin{eqnarray*}
P_0 &=& 0.45 \times 0.94 + 0.55 \times 0.09\\
  &=& 0.4725
\end{eqnarray*}

3. Probability that a 1 was transmitted given that a 1 was received.
\begin{eqnarray*}
P &=& \frac{0.55 \times 0.91}{P_1}\\
  &=& \frac{0.55 \times 0.91}{0.5275}\\
  &=& 0.9488
\end{eqnarray*}

4. Probability that a 0 was transmitted given that a 0 was received.
\begin{eqnarray*}
P &=& \frac{0.45 \times 0.94}{P_0}\\
  &=& \frac{0.45 \times 0.94}{0.4725}\\
  &=& 0.8868
\end{eqnarray*}

5. Probability of an error. (This means either a transmitted 0 was received as a 1, or a transmitted 1 was
received as a 0)
\begin{eqnarray*}
P &=& 0.45\times 0.06 + 0.55 \times 0.09\\
  &=& 0.0765
\end{eqnarray*}

1. P(X = 1) is the probability that two sensors are active in all three time slots:
\begin{eqnarray*}
P(X=1) &=& \frac{1}{4} \times \frac{1}{4} \times \frac{1}{4}\\
       &=& \frac{1}{64}
\end{eqnarray*}

2. X belongs to Bernoulli distribution:
   $$X \sim \text{Bern}(\frac{1}{64})$$.

3. The value of E(X):
\begin{eqnarray*}
E(X) &=& 1 \times P + 0 \times (1-P) \\
     &=& \frac{1}{64}
\end{eqnarray*}

4. The value of Var(X):
\begin{eqnarray*}
Var(X) &=& P \times (1-P) \\
       &=& \frac{1}{64} \times (1 - \frac{1}{64})\\
       &=& \frac{1}{64} \times \frac{64}{64} \\
       &=& \frac{63}{4096}
\end{eqnarray*}

Thus, 
\begin{eqnarray*}
P(X==K) &=& \binom{10}{k} \times 0.4^{k} \times (1-0.4)^{n-k} \\
        &=& \binom{10}{k} \times 0.4^{k} \times (0.6)^{n-k}
\end{eqnarray*}

2. X belongs to Binomial distribution.
   $$X \sim \text{Binom}(10,0.4)$$.

3. Plot the probability mass function (pmf) of X.

```r
n = 10
p = 0.4
y = dbinom(0:10, n,p)
plot(x = 0:10, ylim= (0:1.0), y,  xlab = "k", ylab = "pmf: P(X = k)")
```


4. Plot the cumulative distribution function (cdf) of X.

```r
pmf = dbinom(0:10, 10, 0.4)
print(pmf)
cdf = cumsum(pmf)   # note the use of the R function "cumsum"
print(cdf)
#### Writing a general PMF ####
plot(x=(0:10), cdf, type = 's', ylim = c(0,1), ylab="CDF of X",col=2,xlab="y", main="Cumulative Distribution Funtion");abline(h=0:1,col=4)
```
